```python
    def keys_by_depth(dict_, depth=0, output=None):
        if output is None:
            output = {}
        if not depth in output:
            output[depth] = set()
        for key in dict_:
            output[depth].add(key)
            if isinstance(dict_[key], dict):
                keys_by_depth(dict_[key], depth+1, output)
        return output
```

```python
num=0
s1 = {}
d1 = {}
r1 = {}
for i in data:
    s2 = {}
    d2 = {}    
    r2 = {}
    for j in data[i]:
        s3 = {}
        d3 = {}
        r3 = {}
        for k in data[i][j]:
            s4 = {}
            d4 = {}
            r4 = {}
            for l in data[i][j][k]:
                print(num,end=' ')
                d4['{:.1F}'.format(float(l)/1e-6)] = get_mean_change_in_data(data[i][j][k][l])
                s4['{:.1F}'.format(float(l)/1e-6)] = get_reaction_statistics(data[i][j][k][l])
                r4['{:.1F}'.format(float(l)/1e-6)] = reaction_paths(data[i][j][k][l],max_rows=100)
                num+=1
            d3['{:.1F}'.format(float(k)/1e-6)] = d4
            s3['{:.1F}'.format(float(k)/1e-6)] = s4
            r3['{:.1F}'.format(float(k)/1e-6)] = r4
        d2['{:.1F}'.format(float(j)/1e-6)] = d3
        s2['{:.1F}'.format(float(j)/1e-6)] = s3
        r2['{:.1F}'.format(float(j)/1e-6)] = r3
    d1['{:.1F}'.format(float(i)/1e-6)] = d2
    s1['{:.1F}'.format(float(i)/1e-6)] = s2
    r1['{:.1F}'.format(float(i)/1e-6)] = r2
```
